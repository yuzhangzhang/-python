# python编程基础——1、基本数据结构

## 数据结构（data structure）

- 数据结构是计算机存储、组织数据的方式
- 数据结构是指相互之间存在一种或多种特定关系的数据元素的集合
- 通常情况下，精心选择的数据结构可以带来更高的运行或者存储效率
- 数据结构往往同高效的检索算法和索引技术有关

## 列表 (list)

- 列表最明显的标志就是数据项之间以英文逗号（，）隔开，整个列表用中括号[]包围。
- 列表是最常用的Python数据类型
- 列表中可以混合不同数据类型，不光是数字和字符串，列表中几乎可以存储任意类型数据

***pop 与 remove 对比 :***

pop 移除的是***索引位置对应的元素***，即需要提供`元素位置`，同时***返回该元素值***

remove 移除的是列表中***第一个匹配的元素***，即需提供`要移除的元素值`，并且***没有返回值***

## 元组（tuple）

- 元组的元素***不能修改***
- 元组中元素使用英文逗号（，）隔开，整个元组用小括号()包围。
- 元组索引从 0 开始，也可以使用 -1， -2...

### 1、创建元组

```python
#创建空元组
tup_0 = tuple() 
tup_1 = () 
#创建只包含一个元素的元组，注意要有逗号
tup_2 = (1, ) 
#创建多元素元组
tup_3 = (1, 3, "apple", 66) 
```

### 2、获取元组值

```python
print(tup_3)
```

```python
# 打印元组第0项
print(tup_3[0]) 
```

```python
# 打印元组倒数第2项
print(tup_3[-2])
```

```python
# 打印元组第1项到第2项
print(tup_3[1:3])
```

### 3、元组脚本操作符

```python
# 元组组合
tup_4 = (1,2) + (3,4)
print(tup_4)

# 元组重复
tup_5 = (1,2)*3
print(tup_5)

# 判断数据是否在元组中
print(3 in tup_4)
```

### 4、元组函数

```python
# 元组长度
print(len((1,35,6,0.2,9)))

# 元组最大值
print(max((1,35,6,0.2,9)))

# 元组最小值
print(min((1,35,6,0.2,9)))
```

## 字典（dictionary）

- 字典中的元素为 key:value 对，每对 key:value 用英文逗号隔开，整个字典用大括号{}包围
- 字典元素可变，可存储任意类型对象

### 1、创建字典

```python
# 创建空字典
dict_0 = dict() 
dict_1 = {} 

#创建非空字典
dict_2 = {"name": "张三", "age": 19, 66: "六十六"}
```

### 2、获取字典值

```python
dict_2
```

```python
# 获取字典特定值
dict_2['name']
```

```python
# 获取字典key列表
dict_2.keys()
```

```python
# 获取字典value列表
dict_2.values()
```

### 3、修改字典

```python
dict_2
```

```python
#将键 age 的值更新为30
dict_2["age"]=30
print(dict_2)
```

```python
#给字典dict_2 中增加 "hobby":"eat" 键-值对
dict_2["hobby"] = "eat" 
print(dict_2)
```

```python
#删除字典dict_t 中键为"hobby" 的键-值对
del dict_2["hobby"] 
print(dict_2)
```

### 4、字典函数

```python
#len() 键-值对数目
print(len(dict_2))
```

```python
#list(dic) 以列表形式输出所有键
print(list(dict_2)) 
```

```python
#str(dic)  输出字典可打印的字符串表示
print(str(dict_2))  
```

## 集合 (set)

- 可以使用大括号 { } 或者 set() 函数创建集合，注意：创建一个空集合必须用 set() 而不是 { }，因为 { } 是用来创建一个空字典
- 集合（set）是一个无序的不重复元素序列

### 1、创建集合

```python
basket = { 'orange',  'pear', 'orange', 'banana'}  # 这里演示的是去重功能
print(basket)  
```

### 2、集合的脚本操作符

```python
# 快速判断元素是否在集合内
'orange' in basket  
```

```python
# 集合a中包含而集合b中不包含的元素
a = set('abracadabra')
b = set('alacazam')
print(a,b)
a - b  
```

```python
# 集合a或b中包含的所有元素
a | b   
```

```python
# 集合a和b中都包含了的元素
a & b                              
```

```python
# 不同时包含于a和b的元素
a ^ b                              
```

### 3、集合的基本操作

```python
#添加一个元素
basket.add('apple') 
basket
```

```python
#添加多个元素
basket.update([1,2,3])
basket
```

```python
basket.update([1,2],['dfsf','sdfa'])
basket
```

```python
# 移除元素
basket.remove('apple')
basket
```

```python
basket.discard('pear')
basket
```

***当删除的元素不存在时：***

- ***remove ：*** `会报错`

- ***discard ：*** `不会报错`
